```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get VADER sentiment (positive, negative, neutral)
def get_vader_sentiment(text):
    score = analyzer.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'positive'
    elif score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply VADER sentiment to the review text
df['vader_sentiment'] = df['review'].apply(get_vader_sentiment)

# Adjust sentiment based on rating
def adjust_sentiment_based_on_rating(row):
    if row['rating'] >= 4:
        return 'positive'
    elif row['rating'] <= 2:
        return 'negative'
    else:
        return 'neutral'

df['adjusted_sentiment'] = df.apply(adjust_sentiment_based_on_rating, axis=1)

# Combine VADER sentiment with adjusted sentiment (rating-based)
df['final_sentiment'] = df.apply(lambda row: row['adjusted_sentiment'] if row['vader_sentiment'] == 'neutral' else row['vader_sentiment'], axis=1)
```

```python
#EMOJI TO TEXT
import emoji
df['review']=df['review'].apply(emoji.demojize)
df['title']=df['title'].apply(emoji.demojize)
```

```python
df['review']=df['review'].str.lower()
df['review']=df['review'].str.replace(r"http\S+|www\S+|https\S+", "", regex=True)
df['review']=df['review'].str.replace(r"@\w+|\#", "", regex=True)
df['review']=df['review'].str.replace(r"\d+", "", regex=True)
df['review']=df['review'].str.replace(r"[^\w\s]", "", regex=True)
df['review']=df['review'].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
df['title']=df['title'].str.lower()
df['title']=df['title'].str.replace(r"http\S+|www\S+|https\S+", "", regex=True)
df['title']=df['title'].str.replace(r"@\w+|\#", "", regex=True)
df['title']=df['title'].str.replace(r"\d+", "", regex=True)
df['title']=df['title'].str.replace(r"[^\w\s]", "", regex=True)
df['title']=df['title'].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
#TOKENIZATION
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
df['review']=df['review'].apply(word_tokenize)
df['title']=df['title'].apply(word_tokenize)
```

```python
from spellchecker import SpellChecker
spell = SpellChecker()
def spell_check(text):
  result = []
  for word in text:
    correct_word = spell.correction(word)
    result.append(correct_word)
  return result

df['review'] = df['review'].apply(spell_check)
df['title']=df['title'].apply(spell_check)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
  return [word for word in tokens if word not in stop_words]

df['review'] = df['review'].apply(remove_stopwords)
df['title']=df['title'].apply(remove_stopwords)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_text(tokens):
  return [lemmatizer.lemmatize(word) for word in tokens]
df['review'] = df['review'].apply(lemmatize_text)
df['title']=df['title'].apply(lemmatize_text)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Train-test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, stratify=y_resampled, random_state=42)

# Initialize Logistic Regression with class weight adjustment
model = LogisticRegression(class_weight='balanced', max_iter=1000)

# Train the model on the resampled data
model.fit(X_resampled, y_resampled)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_model.fit(X_resampled, y_resampled)
y_pred_rf = rf_model.predict(X_test)

print("Random Forest:")
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(scale_pos_weight=1, use_label_encoder=False, eval_metric='mlogloss')
xgb_model.fit(X_resampled, y_resampled)
y_pred_xgb = xgb_model.predict(X_test)

print("XGBoost:")
print(classification_report(y_test, y_pred_xgb))
print(confusion_matrix(y_test, y_pred_xgb))
```

```python
from xgboost import XGBClassifier
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Compute class weights
classes = np.unique(y_resampled)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_resampled)
class_weights_dict = {i: w for i, w in zip(classes, class_weights)}

# Set scale_pos_weight indirectly using sample weights
sample_weights = [class_weights_dict[label] for label in y_resampled]

xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb_model.fit(X_resampled, y_resampled, sample_weight=sample_weights)
y_pred_xgb = xgb_model.predict(X_test)

print("XGBoost with Class Weights:")
print(classification_report(y_test, y_pred_xgb, zero_division=0))
```
